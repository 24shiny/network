```python
import networkx as nx
import numpy as np
import itertools as it
import matplotlib.pyplot as plt
```

```python
# a "plain" graph is undirected
G = nx.Graph()

# give each a node a 'name', which is a letter in this case.
G.add_node('a')

# the add_nodes_from method allows adding nodes from a sequence, in this case a list
nodes_to_add = ['b', 'c', 'd']
G.add_nodes_from(nodes_to_add)

# add edge from 'a' to 'b'
# since this graph is undirected, the order doesn't matter here
G.add_edge('a', 'b')

# just like add_nodes_from, we can add edges from a sequence
# edges should be specified as 2-tuples
edges_to_add = [('a', 'c'), ('b', 'c'), ('c', 'd')]
G.add_edges_from(edges_to_add)

# draw the graph
nx.draw(G, with_labels=True)
```

```python
# There are many optional arguments to the draw function to customize the appearance.
nx.draw(G,
        with_labels=True,
        node_color='lightpink',
        node_size=500,
        font_color='white',
        font_size=12,
        )
```

```python
# check basic information of the graph

# List all of the nodes
print(type(G.nodes()))
print(G.nodes())

# List all of the edges
print(type(G.edges()))
print(G.edges())
```

```python
# NodeView and EdgeView objects have iterators, so we can use them in for loops:
# to this end, the itertools it is imported

for node in G.nodes:
    print(node)
    
freqlist = []
for node in G.nodes:
    freqlist.append(G.degree(node))
print(freqlist)
```

```python
for edge in G.edges:
    print(edge)
```

```python
# check the number of nodes and edges
print(G.number_of_nodes())
print(G.number_of_edges())
```

```python
# list of neighbors of node 'b'
list(G.neighbors('b'))
```

```python
for neighbor in G.neighbors('b'):
    print(neighbor)
```

```python
# use the list constructor to make a list from an iterator
l = list(G.neighbors('b'))
l[0]
```

```python
# a couple of examples of NetworkX module functions that provide information about a graph:
print(nx.is_tree(G))
print(nx.is_connected(G))
```

```python
# Node and edge existence
print(G.has_node('a'))
print(G.has_node('x'))

# Additionally, the loop syntax used above: 
# for n in G.nodes suggests another way we can check if a node is in a graph
print('d' in G.nodes)

# Likewise we can check if two nodes are connected by an edge:
print(G.has_edge('a', 'b'))
print(('c', 'd') in G.edges)
```

```python
# node degree
print(len(list(G.neighbors('a'))))
print(G.degree('a'))
```
