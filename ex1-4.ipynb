```python
import networkx as nx
import numpy as np
import itertools as it
import matplotlib.pyplot as plt
```

# EXERCISE 4
Write a function that gets a Graph object as input and returns the average degree and the standard deviation -
Test your function with the karate network.
Plot the degree distribution of the karate network (4 pts)

```python
K = nx.karate_club_graph()
pos = nx.circular_layout(K)
nx.draw_networkx(K, pos=pos, with_labels=True, node_color='skyblue')
```

```python
def get_avg_std(G):
    degree = np.array(list(dict(G.degree()).values()))
    print(np.round(np.mean(degree),3), np.round(np.std(degree),3))

    l_k = list(dict(G.degree()).values())
    s_k = set(l_k)
    
    dict_k_frequency = {}
    for k in s_k:
        dict_k_frequency[k] = l_k.count(k)
    
    plt.bar(dict_k_frequency.keys(), dict_k_frequency.values(),width=.4)
    plt.xlabel('degree',fontsize = 16)
    plt.ylabel('frequency',fontsize = 16)
    plt.show()
```

```python
get_avg_std(K)
```
